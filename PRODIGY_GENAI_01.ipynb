```python
# ========================
# STEP 1 — Install Compatible Versions
# ========================
!pip install --upgrade torch==2.1.0 torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!pip install --upgrade transformers==4.44.2 datasets==2.21.0 accelerate==0.33.0 matplotlib

# ========================
# STEP 2 — GPU Check
# ========================
import torch
device_name = torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU"
print(f"Using device: {device_name}")

# ========================
# STEP 3 — Import Libraries
# ========================
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments, TrainerCallback
from datasets import load_dataset
from google.colab import files
import matplotlib.pyplot as plt

# ========================
# STEP 4 — Upload Dataset
# ========================
uploaded = files.upload()
dataset_path = list(uploaded.keys())[0]

# ========================
# STEP 5 — Load Model & Tokenizer
# ========================
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id

# ========================
# STEP 6 — Load and Split Dataset
# ========================
dataset = load_dataset('text', data_files={'data': dataset_path})["data"]
dataset = dataset.train_test_split(test_size=0.1, seed=42)

# ========================
# STEP 7 — Tokenize Dataset
# ========================
def tokenize_function(examples):
    encodings = tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=128
    )
    encodings["labels"] = encodings["input_ids"].copy()
    return encodings

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# ========================
# STEP 8 — Callback for Live Loss Tracking
# ========================
train_losses, eval_losses = [], []

class TextGenerationCallback(TrainerCallback):
    def __init__(self, tokenizer, sample_prompt="Once upon a time"):
        self.tokenizer = tokenizer
        self.sample_prompt = sample_prompt

    def on_log(self, args, state, control, logs=None, **kwargs):
        if "loss" in logs:
            train_losses.append(logs["loss"])
        if "eval_loss" in logs:
            eval_losses.append(logs["eval_loss"])

    def on_epoch_end(self, args, state, control, **kwargs):
        print(f"\n--- Sample Generation After Epoch {state.epoch} ---")
        inputs = self.tokenizer.encode(self.sample_prompt, return_tensors="pt").to(kwargs['model'].device)
        outputs = kwargs['model'].generate(
            inputs,
            max_length=50,
            num_return_sequences=1,
            temperature=0.7,
            top_k=50,
            top_p=0.95
        )
        print(self.tokenizer.decode(outputs[0], skip_special_tokens=True))
        print("------------------------------------\n")

# ========================
# STEP 9 — Training Arguments
# ========================
training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",
    evaluation_strategy="epoch",  # now works
    learning_rate=5e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    weight_decay=0.01,
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=50,
    fp16=torch.cuda.is_available(),
    report_to="none"
)

# ========================
# STEP 10 — Trainer Setup
# ========================
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer,
    callbacks=[TextGenerationCallback(tokenizer)]
)

# ========================
# STEP 11 — Train
# ========================
trainer.train()

# ========================
# STEP 12 — Save Model
# ========================
trainer.save_model("./gpt2-finetuned")
tokenizer.save_pretrained("./gpt2-finetuned")

# ========================
# STEP 13 — Final Generation
# ========================
prompt = "The future of artificial intelligence is"
inputs = tokenizer.encode(prompt, return_tensors="pt").to(model.device)
outputs = model.generate(
    inputs,
    max_length=60,
    num_return_sequences=1,
    temperature=0.7,
    top_k=50,
    top_p=0.95
)
print("\nFinal Generated Text:\n", tokenizer.decode(outputs[0], skip_special_tokens=True))

# ========================
# STEP 14 — Plot Loss Graph
# ========================
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label="Training Loss")
plt.plot(eval_losses, label="Validation Loss")
plt.xlabel("Logging Steps")
plt.ylabel("Loss")
plt.title("GPT-2 Fine-tuning Loss Curve")
plt.legend()
plt.grid(True)

graph_path = "loss_curve.png"
plt.savefig(graph_path, dpi=300)
plt.show()

# ========================
# STEP 15 — Download Graph
# ========================
files.download(graph_path)
```
